# Linear Regression Using Closed-Form Solution vs scikit-learn

### 1\- Introduction
In this notebook, we implement **linear regression** using the **closed-form solution (normal equation)** and compare it with **scikit-learn's `LinearRegression`**.  
We will evaluate both models using the **R² score**.

---

```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

### 2\- Dataset
We use the **California Housing dataset** from `sklearn.datasets`.  

- `X_train` – feature matrix  
- `y_train` – target vector (house prices)  

This dataset has **8 numerical features** per example and **over 20,000 samples**.

---

```python
from sklearn.datasets import fetch_california_housing
X_train, y_train = fetch_california_housing(return_X_y=True)
```

### 3\- Linear Regression Closed-Form
The **closed-form solution** computes the optimal weights analytically using the **normal equation**:
$$
\mathbf{w} = (X^T X)^{-1} X^T y
$$
Where:

- $X$ = feature matrix (with a column of ones for the intercept)  
- $y$ = target vector  
- $\mathbf{w}$ = weight vector (first element = intercept)

**Key points:**

- We add a **column of ones** to the feature matrix for the intercept.  
- Matrix multiplication `X^T X` is used to capture the covariance between features.  
- `np.linalg.inv()` computes the **inverse** needed for the normal equation.  
- This method gives an **exact solution** for the weights.

---


```python
class LinearRegressionClosedForm:
    def __init__(self):
        self.coef_ = None
        self.intercept_ = 0.0
    
    def fit(self, X_train, y_train):
        X_train = np.array(X_train)
        y_train = np.array(y_train)

        X_train = np.c_[np.ones((X_train.shape[0], 1)), X_train]

        A = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
        self.intercept_ = A[0]
        self.coef_ = A[1:]

    def predict(self, X):
        X = np.array(X)

        return X @ self.coef_ + self.intercept_
```

### 4\- Predictions
Once both models are trained, we predict the target values for the training dataset:

- `y_pred_` → predictions from the **closed-form model**  
- `y_pred_sk` → predictions from the **scikit-learn model**

This allows us to compare how well both approaches fit the data.

---

```python
my_model = LinearRegressionClosedForm()
sk_model = LinearRegression()

my_model.fit(X_train, y_train)
sk_model.fit(X_train, y_train)

y_pred_ = my_model.predict(X_train)
y_pred_sk = sk_model.predict(X_train)


```

### 5\- Evaluation with R² Score
The **R² score** measures how well the model explains the variance of the target:

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

- A score **close to 1** indicates the model explains most of the variance.  
- Comparing the R² scores of both models confirms that our **closed-form solution is correct**.
